```python
import pandas as pd
import numpy as np
import ast
```

We will create tag based on relevent features. Like genres play a big role, to get similiarity of two users. Budget is not as good for this purpose. lets say i liked Inception(high budget movie) but i need not like other high budget movie like DC movies. based on this situation lest filterout the relevent features to make tag

```python
# genres
# id        -- Will be usefull to fetch posters of movie
# keywords
# title
# overview
# cast

movies = movies[['movie_id', 'title', 'overview','genres','keywords','cast']]
```

```python
movies.head(2)
```

```python
# missing data
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# fetching genres in list format form this dictonary format

def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['keywords'] =movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(2)
```

```python
# converting overview sting column to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x] )
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x] )
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x] )
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']
```

```python
movies.head(2)
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
# convert in lowwer case
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
# applying stem to convert this string : [ love, loving, loved]
# to this : [love, love, love]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Titanic')
```
